```python
import pandas as pd
import numpy as np

from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier
```

```python
X,y = make_classification(n_samples = 10, n_classes = 2, n_features = 2, n_informative = 2,n_redundant = 0,random_state = 42)
```

```python
clf = DecisionTreeClassifier()
clf.fit(X,y)
```

```python
from sklearn.tree import plot_tree
plot_tree(clf)
```

```python
clf.feature_importances_
```

```python
#importance using random forest is average of feature importance obtained using Decision tree
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()

rfc.fit(X,y)
```

```python
rfc.feature_importances_
```
