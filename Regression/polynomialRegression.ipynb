```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Position_Salaries.csv')
data
```

```python
#We will consider only two columns i.e., Level and Salary as it Level is encoded form of Position
#Extracting independent(X) and dependent(y) variable
X = data.iloc[:, 1: 2]
y = data.iloc[:,-1:]
```

```python
print(X)
```

```python
print(y)
```

```python
#We will not split data into train and test as there is no enough data.
#The above dataset is non linear so we will have to apply Polynomial regresson for better accuracy
#We will also apply lineae regression to this dataset for comparison.
```

```python
#Usinf simple linear model
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X,y)
```

```python
#using polynomial regression model
from sklearn.preprocessing import PolynomialFeatures

poly_regs = PolynomialFeatures(degree=4)
X_poly = poly_regs.fit_transform(X)
```

```python
#fitting to linear model
poly_lin_regs = LinearRegression()
poly_lin_regs.fit(X_poly, y)

```

```python
#Visualizing the results of simple linear model
plt.scatter(X,y)
plt.plot(X,regressor.predict(X), color='red')
plt.title('Level Vs Salary (Simple Linear Model)')
plt.xlabel('Level')
plt.ylabel("Salary")
```

```python
#Visualizing the results of polynomial regression model
plt.scatter(X,y)
plt.plot(X,lin_regs.predict(X_poly), color='red')
plt.title('Level Vs Salary (Polynimial Regression Model)\n Degree = 2')
plt.xlabel('Level')
plt.ylabel("Salary")
```

```python
# Now we will see result by increasing degree
plt.scatter(X,y)
plt.plot(X,lin_regs.predict(X_poly), color='red')
plt.title('Level Vs Salary (Polynimial Regression Model)\n Degree = 3')
plt.xlabel('Level')
plt.ylabel("Salary")
```

```python
# Now we will see results fro degree 4
#Visualizing the results of polynomial regression model
plt.scatter(X,y)
plt.plot(X,lin_regs.predict(X_poly), color='red')
plt.title('Level Vs Salary (Polynimial Regression Model)\n Degree = 4')
plt.xlabel('Level')
plt.ylabel("Salary")
```

```python
#In the above graph we can cearly see that as we go on increasig the degree of polynoial then the accuracy will also increase
#Now we will compare the results of simple linear regression model and polynomial regression model
print('Simple linear model for Level 6 : ',regressor.predict([[6]]))
print('Polynomial regression model for level 6 : ',poly_lin_regs.predict(poly_regs.fit_transform([[6]])))
print("From the above comparison we can see that the output given by Polynomial regression model is much closer to original values than that of Simple lineae regression model")
```
